```python
import zipfile
import os
import glob
import numpy as np
from tensorflow.keras.preprocessing import image

# step 1:- Unzip the uploaded file
with zipfile.ZipFile("/content/Data.zip", "r") as zip_ref:
    zip_ref.extractall("/content/Data")

# step 2:- Check folder structure
print("Folders inside /content/Data.zip:", os.listdir("/content/Data"))
```

```python
# Correct base path where dataset is extracted
base_dir = '/content/Animal/Animal'

# Step 3:- Get image file paths

fruit_paths = glob.glob(os.path.join(base_dir, 'Fruit', '*'))   # All fruit image paths
vegetable_paths = glob.glob(os.path.join(base_dir, 'Vegetable', '*'))   # All vegetable image paths

# Print number of images found
print('Found', len(fruit_paths), 'fruit images')
print('Found', len(vegetable_paths), 'vegetable images')
```

```python
#step 4:- Create empty lists
# -----------------------------

data = [] #X images
labels = [] #y name
```

```python

# Step 5:- Load and preprocess Fruit images


for path in fruit_paths:

    # Load image and resize to 150x150
    img = image.load_img(path, target_size=(150, 150))

    # Convert image to numpy array
    img_array = image.img_to_array(img)

    # Normalize pixel values from [0,255] → [0,1]
    img_array = img_array / 255.0

    # Store processed image
    data.append(img_array)

    # Assign label 0 for cat (target)
    labels.append(0)


```

```python
# step 6:- Load and preprocess Vegetable images


for path in vegetable_paths:

    # Load image and resize to 150x150
    img = image.load_img(path, target_size=(150, 150))

    # Convert image to numpy array
    img_array = image.img_to_array(img)

    # Normalize pixel values
    img_array = img_array / 255.0

    # Store processed image
    data.append(img_array)

    # Assign label 1 for dog
    labels.append(1)
```

```python
# step 7:-  CNN models require numpy arrays
x = np.array(data)      # Image data
y = np.array(labels)
```

```python
# Step 8:- Check dataset structure
print("Data shape:", x.shape)   # (num_images, 150, 150, 3)
print(" Labels:", y)
```

```python
#step 9:- Create an empty list to store labels
labels = []

# Loop through all fruit image paths
for path in fruit_paths:
    # Add label 'fruit' for each cat image
    labels.append('fruit')

# Loop through all vegetable image paths
for path in vegetable_paths:
    # Add label 'vegetable' for each dog image
    labels.append('vegetable')

# Example output
# This will print labels corresponding to the images
print(labels)  # ['cat', 'cat', 'dog', 'dog']
```

```python
# step 10:-  Convert labels to one-hot
from tensorflow.keras import models, layers
from tensorflow.keras.utils import to_categorical

y_cat = to_categorical(y)
y_cat

```

```python
# step 11:- Build CNN
model = models.Sequential([
    layers.Conv2D(64, (3,3), activation='relu', input_shape=(150, 150, 3)),
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(32, (3,3), activation='relu'),
    layers.Conv2D(32, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),

    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(2, activation='softmax')
])

# step 12:- Compile
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# step 13:- Train
model.fit(x, y_cat, epochs=10, batch_size=32)
```
