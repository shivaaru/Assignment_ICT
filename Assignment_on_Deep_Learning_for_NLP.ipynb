```python
# Import Pandas
import pandas as pd

#Load Data
data = pd.read_csv('/content/judge-1377884607_tweet_product_company.csv',encoding='ISO-8859-1')
data.head(3)
```

```python
df=data.drop("emotion_in_tweet_is_directed_at",axis=1) #drop the id column
```

```python
import pandas as pd
import re # for replacing
import nltk # natural language tool kit
from nltk.corpus import stopwords # Importing stopwords

nltk.download('stopwords') # downloading stop words
stop_words = set(stopwords.words('english'))

def clean_text(text):
    text = str(text).lower() # converting to Lowercase
    text = re.sub(r'http\S+|@\w+|#\w+', '', text) # Removing http,@,# char and replace it with ''
    text = re.sub(r'[^a-z\s]', '', text) # only keep letters and replace all others with ''
    words = text.split() # splitting the words
    words = [w for w in words if w not in stop_words] # will not include stopwords
    return " ".join(words)

df['clean_text'] = df['tweet_text'].apply(clean_text)
```

```python
# Tokenization + Padding
from tensorflow.keras.preprocessing.text import Tokenizer #imported Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences #imported pad_Sequence

tokenizer = Tokenizer(num_words=5000)
tokenizer.fit_on_texts(df['clean_text'])

seqs = tokenizer.sequences_to_texts(df['clean_text']) # Text to seq
X_pad = pad_sequences(seqs, maxlen=100) # maxlength for each tweet
```

```python
# 1.RNN
from tensorflow.keras.models import Sequential # Importing Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense # Importing RNN,Simple Rnn,Dense

model = Sequential([
    Embedding(5000, 128),
    SimpleRNN(64),
    Dense(13, activation='softmax')
])
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer# Importing Tokenizer and Pad_sequence
from tensorflow.keras.preprocessing.sequence import pad_sequences

tokenizer = Tokenizer(num_words=5000) # keeps only top 5000 frequent words
tokenizer.fit_on_texts(df['clean_text'])

X = tokenizer.texts_to_sequences(df['clean_text'])# Text to Sequence
X = pad_sequences(X, maxlen=100)
```

```python
from sklearn.preprocessing import LabelEncoder# Importing Label encoder

le = LabelEncoder()
y = le.fit_transform(df['is_there_an_emotion_directed_at_a_brand_or_product']) # Encoding the y

print(y[:5])
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, y, epochs=5)
```

```python
new_text = ["I am very happy today"]

# step 1: tokenize
seq = tokenizer.texts_to_sequences(new_text)

# step 2: pad
padded = pad_sequences(seq, maxlen=100)

# step 3: predict
pred = model.predict(padded)

print(pred)
```

```python
import numpy as np

pred_class = np.argmax(pred)
print(pred_class)
```

```python
label = le.inverse_transform([pred_class])
print(label[0])
```

```python
#2.lstm

from tensorflow.keras.layers import LSTM

model = Sequential([
    Embedding(5000, 128),
    LSTM(64),
    Dense(13, activation='softmax')
])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,      # 20% test data
    random_state=42     # reproducibility
)
```

```python
model.compile(
    loss='sparse_categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

model.fit(X_train, y_train, epochs=5, validation_data=(X_test, y_test))
```

```python
new_text = ["I don't Like this product.I get's hang every time"]

seq = tokenizer.texts_to_sequences(new_text)
padded = pad_sequences(seq, maxlen=100)

pred = model.predict(padded)

import numpy as np
pred_class = np.argmax(pred)

print("Predicted Class:", pred_class)
label = le.inverse_transform([pred_class])
print(label[0])
```
