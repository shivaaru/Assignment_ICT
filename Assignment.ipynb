# **2. Electricity Bill Generator**


```python
Units_consumed = eval(input("Enter the Units consumed: "))
if(Units_consumed > 0) and (Units_consumed <= 200):
   if(Units_consumed <= 100):
        less_100=Units_consumed*5
        print("Bill amount : ₹",less_100)
   else:
        less_100 =100*5
        less_200 = (Units_consumed-100)*7
        print("Bill amount : ₹",less_200+less_100)
elif(Units_consumed > 200):
    morethan_200= (Units_consumed - 200)*10
    combined =(100*5)+(100*7)
    print("Bill amount : ₹",morethan_200 + combined)
else:
  print("Incorrect value")
```

# **6. Sum of even numbers**

# **7. Dictionary – Student Marks Summary**

```python
students = {}
total_marks=0
no_of_students=3
for _ in range(no_of_students):
  name=input("enter name:")
  marks=int(input("Enter Mark:"))
  students[name]=marks
  total_marks+=marks
avg1=total_marks/no_of_students
avg=f"{avg1:.1f}"
print("Average Marks:",avg)
```

# **8. Functions – Check Prime**
