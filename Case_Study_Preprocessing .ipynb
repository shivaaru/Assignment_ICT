# **CASE STUDY 01**  :- **Preprocessing**

 **1. Load the dataset in Python environment**  

```python
# Imported pandas,seaborn,numpy and statistics
import pandas as pd
import seaborn as sea
import numpy as np
import statistics as st
```

```python
# Loaded the data
data = pd.read_csv('/content/train_loan.csv')
data
```

**Missing value handling**

```python
# Getting the info to check missing values
data.info()
# Here multiple columns have missing values
```

```python
# Checked for Duplicate values in data
print("Duplicates count in the given data:",data.duplicated().sum())
# In this data there are duplicates rows
```

```python
# Check if there is any Null value
data.isnull().mean()*100

# Here the range is between 1- 40 so we can use fillna()
```

```python
data.describe()
```

```python
# First let's solve the null value in gender using 'mode', since it is a object
Gender = data['Gender']
st.mode(data['Gender'])
data['Gender'].fillna(st.mode(data['Gender']),inplace=True)
```

```python
# Checking the issue with gender is solved
data.info()
data.isnull().mean()*100
#Now geneder has no missing values
```

```python
# Since married column has less than 1% of missing values we can use dropna()

df1 = data.dropna(subset=['Married'])
df1.isnull().mean()*100
```

```python
# Next we can solve the issue of Dependents with fillna() using mode as it is object
st.mode(df1['Dependents'])
df1['Dependents'].fillna(st.mode(df1['Dependents']),inplace=True)
```

```python
df1.isnull().mean()*100
```

```python
# Next we can solve the issue of Self_Employed with fillna() using mode as it is object
st.mode(df1['Self_Employed'])
df1['Self_Employed'].fillna(st.mode(df1['Self_Employed']),inplace=True)
df1.isnull().mean()*100
```

```python
# LoanAmount is  float, so we can use median with fillna()
# Skewness of LoanAmont
print("Skewness of LoanAmont:-",df1['LoanAmount'].skew())
df1['LoanAmount'].fillna(st.median(df1['LoanAmount']),inplace=True)
df1.isnull().mean()*100
```

```python
# Loan_Amount_Term is  float, so we can use median with fillna()
# Skewness of Loan_Amount_Term
print("Skewness of Loan_Amount_Term:-",df1['Loan_Amount_Term'].skew())
df1['Loan_Amount_Term'].fillna(st.median(df1['Loan_Amount_Term']),inplace=True)
df1.isnull().mean()*100
```

```python
# Credit_History is  float, so we can use median with fillna()
# Skewness of Credit_History
print("Skewness of Credit_History:-",df1['Credit_History'].skew())
df1['Credit_History'].fillna(st.median(df1['Credit_History']),inplace=True)
df1.isnull().mean()*100
```

```python
#Since Gender,Married, Property_Area,Loan_Status has 2 cardinality we can use One-Hot Encoding and also it is nominal
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder
le = LabelEncoder()
df1['Gender'] = le.fit_transform(df1['Gender'])
df1['Married'] = le.fit_transform(df1['Married'])
df1['Property_Area'] = le.fit_transform(df1['Property_Area'])
df1['Loan_Status']= le.fit_transform(df1['Loan_Status'])
```

```python
#Everything has changed to digits
df1
```

```python
# Education is ordinal data
oe = OrdinalEncoder(categories=[['Not Graduate','Graduate']]) # In asc order
df1['Education']=oe.fit_transform(df1[['Education']])
```
